# Seaborn을 이용한 데이터시각화 학습

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### seaborn에서 제공하는 flights 데이터 셋

```python
flights = sns.load_dataset('flights')
```

## Key 확인

```python
flights.keys()
```

## DataFrame 형태로 만들기

```python
df = pd.DataFrame(flights)
df
```

# Plot Style 알아보기

## Seaborn Set Theme

```python
# sns.set_style, sns.set_plette, sns.set_context 등으로 접근할 수 있음
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.barplot(x='year', y='passengers', data=df)
```

## Seaborn Set Style

```python
# style : dict, None, or one of {darkgrid, whitegrid, dark, white, ticks}
# rc : dict, optional

# 사용 예시
# >>> set_style("whitegrid")
# >>> set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})

sns.set_style('whitegrid')
sns.barplot(x='year', y='passengers', data=df)
```

## Seaborn Spin

```python
# sns.despine() : 테두리 제거

# 속성 및 Default 
# fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False

# 사용 예시
# >>> sns.despine(top=True, bottom=True)
sns.barplot(x='year', y='passengers', data=df)
sns.despine(left=True, bottom=True)
```

# Seaborn Set Context

```python
# 속성 및 기본값
# context='notebook', font_scale=1, rc=None

# context 종류
# 'paper' - 'notebook' - 'talk' - 'poster' / 가장 작은 scale - 기본 scale - 조금 더 큰 scale - 가장 큰 scale

plt.figure(figsize=(12, 5))

sns.set_context('poster')
sns.barplot(x='year', y='passengers', data=df)

plt.show()
```

```python
sns.set_context('paper')
sns.barplot(x='year', y='passengers', data=df)
```

## Seaborn Set Palette

```python
# 속성 및 기본값
# palette, n_colors=None, desat=None, color_codes=False / desat : 각 색상의 채도
# palette= deep, muted, pastel, bright, dark, colorblind

sns.set_context('notebook')
sns.set_palette('pastel', 12)
sns.barplot(x='year', y='passengers', data=df)
```

## Seaborn Set Palette 색상 직접 지정하기

```python
colors = ["#FF73B8", "#FFFF36", "#FF73B8", "#FFFF36", "#FF73B8", "#FFFF36", "#FF73B8", "#FFFF36", "#FF73B8", "#FFFF36", "#FF73B8", "#FFFF36"]
sns.set_palette(palette=sns.color_palette(colors))
sns.barplot(x='year', y='passengers', data=df)
```

## Seaborn Font_scale

```python
plt.figure(figsize=(12, 5))

# scale보다는 font_scale이 우선 순위임
sns.set_context('poster', font_scale=.5)
sns.barplot(x='year', y='passengers', data=df)

plt.show()
```

## Seaborn Rc

```python
# 사용할 수 있는 RC 파라미터
# {'axes.labelsize': 17.6,
#  'axes.titlesize': 19.2,
#  'font.size': 19.2,
#  'grid.linewidth': 1.6,
#  'legend.fontsize': 16.0,
#  'lines.linewidth': 2.8,
#  'lines.markeredgewidth': 0.0,
#  'lines.markersize': 11.2,
#  'patch.linewidth': 0.48,
#  'xtick.labelsize': 16.0,
#  'xtick.major.pad': 11.2,
#  'xtick.major.width': 1.6,
#  'xtick.minor.width': 0.8,
#  'ytick.labelsize': 16.0,
#  'ytick.major.pad': 11.2,
#  'ytick.major.width': 1.6,
#  'ytick.minor.width': 0.8}

plt.figure(figsize=(12, 5))

sns.set_context('poster', font_scale=.5, rc={"grid.linewidth": 5})
sns.barplot(x='year', y='passengers', data=df)

plt.show()
```

# Barplot

```python
sns.barplot(x='year', y='passengers', data=df)
```

## Barplot hue 속성

```python
plt.figure(figsize=(20, 4))

# hue(유형) 속성 : 월별, 성별, 일별 등으로 표현할 때 사용
sns.barplot(x='year', y='passengers', hue='month', data=df)
plt.show()
```

## Barplot order 속성

```python
# 사용 예시
# >>> sns.barplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"])

order = df.year.sort_values(ascending=False).unique().tolist()
print(order)

sns.barplot(x='year', y='passengers', data=df, order=order)
```
